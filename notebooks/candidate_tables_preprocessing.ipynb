# Candidate Tables Preprocessing

To evaluate the model performance we also include some tables from other sources than wikipedia.

```python
import pandas as pd
import numpy as np
import names
```

## Titanic

Link: https://www.kaggle.com/c/titanic

```python
df = pd.read_csv('/home/mikexydas/Downloads/train.csv')
df = df.drop(['SibSp', 'Parch', 'Pclass', 'Embarked', 'Cabin'], axis=1)
df = df.dropna()

# We replace the column name with new random names since the current column includes family info.
df['Name'] = df['Name'].apply(lambda x: names.get_full_name())

df['Survived'] = df['Survived'].apply(lambda x: "Survived" if x==0 else "Died")
df['Age'] = df['Age'].apply(np.int64)
```

```python
df.to_csv('../storage/datasets/random_datasets/titanic.csv', index=False)
display(df)
```

## Student Performance

Link: https://www.kaggle.com/spscientist/students-performance-in-exams

```python
df = pd.read_csv('/home/mikexydas/Downloads/archive/StudentsPerformance.csv')
df = df.rename(columns = {"race/ethnicity":"ethnicity", 
                          "test preparation course": "test_preparation_course",
                          "math score": "math_score",
                          "reading score": "reading_score",
                          "writing score": "writing_score",
                          "parental level of education": "parental_level_of_education"})
df.insert(0, "student_id", df.index)
df = df[:300]
```

```python
df.to_csv('../storage/datasets/random_datasets/student_performance.csv', index=False)
display(df)
```

## Netflix Shows

Link: https://www.kaggle.com/shivamb/netflix-shows

```python
df = pd.read_csv('/home/mikexydas/Downloads/archive/netflix_titles.csv')
df = df.drop(['cast', 'listed_in', 'description'], axis=1)
df = df.dropna()
df.insert(0, "student_id", df.index)
df = df.sample(n=300)
```

```python
df.to_csv('../storage/datasets/random_datasets/netflix_shows.csv', index=False)
display(df)
```
